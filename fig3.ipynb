```python
import matplotlib.pyplot as plt
import math
import random
```

```python
# モデルのパラメータ
b0 = 1*10**(-7)
b1 = 2*b0
c = 1.3*10
B = 1
gamma=200
a0=0.01
a1=1.75*a0

# 初期値
x = 1*10**8

# 目安
z2=(c/x - b0) / (b1 - b0)
y2 =( ((1-z2)/z2) *B*(1+gamma*z2**2) - a0)/a1

print(y2, z2)
```

```python
z0 = (c - b0*x)/((b1-b0)*x)
sigma = B*gamma/((b1-b0)*x)
lamda = a0/(B*gamma)
ky = a1/(B*gamma)

f = ((1- z0 )/z0)*(z0*z0+1/gamma)
y0 = f -lamda

print(z0, gamma, sigma, lamda, f, y0)
```

```python
# 初期値
# x = 1*10**8
# y = 2500*ky*random.random()
# z = random.random()
y = 2500*ky*random.random()
z = random.random()
t = 0.0

# 時間の設定
dt = 0.01
tEnd = 100
iEnd = int(tEnd/dt)

t_list = [t]
y_list = [y]
z_list = [z]

for i in range(1, iEnd):
    t = dt*i
    y_new = y + dt*(z-z0)*y
    z_new = z + dt*sigma*z*(((1- z )/z)*(z*z+1/gamma) - lamda - y)
    y = y_new
    z = z_new

    t_list.append(t)
    y_list.append(y)
    z_list.append(z)
```

```python
# 時間発展のプロット
plt.plot(t_list, y_list)
plt.savefig("fig3b_1.png", format="png", dpi=300)
```

```python
# 時間発展のプロット
plt.plot(t_list, z_list)
plt.legend(["infect"],loc="upper right")
```

```python
# lamda
lamda1 = 0
lamda2 = 0.3
dlamda = 10**((-3))
length = int((lamda2-lamda1)/dlamda)

lamda_list=[]
min1_list=[]
max1_list=[]
min2_list=[]
max2_list=[]

for k in range(1,length):
    y1_list = []
    z1_list = []
    lamda = lamda1+dlamda*k
    # epsilon = 100
    t=0
    y1 = 2500*ky*random.random()
    z1 =  random.random()
    lamda_list.append(lamda)
    for l in range(0, iEnd):
        t = dt*l
        y1_new = y1 + dt*(z1-z0)*y1
        z1_new = z1 + dt*sigma*z1*(((1- z1 )/z1)*(z1*z1+1/gamma) - lamda - y1)
        y1 = y1_new
        z1 = z1_new

        if t>50:
            y1_list.append(y1)
            z1_list.append(z1)

    z1_min=min(z1_list)
    z1_max=max(z1_list)
    y1_min=min(y1_list)
    y1_max=max(y1_list)

    min1_list.append(z1_min)
    max1_list.append(z1_max)
    min2_list.append(y1_min)
    max2_list.append(y1_max)
```

```python
plt.plot(lamda_list, min1_list,'.',color='black')
plt.plot(lamda_list, max1_list,'.',color='red')
plt.savefig("fig3a_1.png", format="png", dpi=300)
```
