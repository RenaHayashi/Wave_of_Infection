```python
import matplotlib.pyplot as plt
import math
import random
```

```python
# モデルのパラメータ
b0 = 1*10**(-7)
b1 = 2*b0
c = 1.3*10
B = 1
gamma=200
a0=0.01
a1=1.75*a0

# 初期値
x = 1*10**8

# 目安
z2=(c/x - b0) / (b1 - b0)
y2 =( ((1-z2)/z2) *B*(1+gamma*z2**2) - a0)/a1

print(y2, z2)
```

```python
z0 = (c - b0*x)/((b1-b0)*x)
sigma = B*gamma/((b1-b0)*x)
lamda = a0/(B*gamma)
ky = a1/(B*gamma)

print(z0, sigma,lamda)
```

```python
# 初期値
# x = 1*10**8
# y = 2500*ky*random.random()
# z = random.random()
y = 2500*ky*random.random()
z = random.random()
t = 0.0

# 時間の設定
dt = 0.01
tEnd = 100
iEnd = int(tEnd/dt)

t_list = [t]
y_list = [y]
z_list = [z]

for i in range(1, iEnd):
    t = dt*i
    y_new = y + dt*(z-z0)*y
    z_new = z + dt*sigma*z*(((1- z )/z)*(z*z+1/gamma) - lamda - y)
    y = y_new
    z = z_new

    t_list.append(t)
    y_list.append(y)
    z_list.append(z)
```

```python
# 時間発展のプロット
plt.plot(t_list, y_list)
```

```python
# 時間発展のプロット
plt.plot(t_list, z_list)
```

```python
#関数f(x)
b0 = 1*10**(-7)
b1 = 2*b0
gamma=200
x = 1*10**8

c=10.1
z0 = (c - b0*x)/((b1-b0)*x)
# z0=1
lamda =((1- z0 )/z0)*(z0*z0+1/gamma)
# 時間の設定
dc = 0.1
cEnd = 25
iEnd = int((cEnd-c)/dc)

c_list = [c]
lamda_list = [lamda]

for i in range(2, iEnd):
    c = 10+dc*i
    zz = (c - b0*x)/((b1-b0)*x)
    lamda =((1- zz )/zz)*(zz*zz+1/gamma)
    
    c_list.append(c)
    lamda_list.append(lamda)
```

```python
plt.plot(c_list, lamda_list, color="black")
plt.xlim(10, 22)
plt.ylim(0, 0.6)
plt.savefig("fig7a.png", format="png", dpi=300)
```
